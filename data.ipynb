```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("car_data.csv")
data.head()
```

```python
data.describe() 
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# Determine the range of manufacturing years
print(f"Vehicles are present from {data['Year'].min()} to {data['Year'].max()}.")
```

```python
# Determine the lowest price
print(f"The lowest price to which a vehicle is sold is {data['Selling_Price'].min()}.")
```

```python
# Determine the highest price
print(f"The highest price to which a vehicle is sold is {data['Selling_Price'].max()}.")
```

```python
# Determine the number of records
print(f"There are {len(data)} records in this data.")
```

```python
# Check for missing records
print(f"Are there any missing records in this data? {'Yes' if data.isnull().values.any() else 'No'}.")
```

```python
# Determine the number of different vehicles
print(f"There are {data['Car_Name'].nunique()} different vehicles in this data.")
```

```python
# Determine the most sold vehicle
most_sold_vehicle = data['Car_Name'].value_counts().idxmax()
print(f"The most sold vehicle in this data is {most_sold_vehicle}.")
```

```python
# Check for CNG vehicles
cng_vehicles = data[data['Fuel_Type'] == 'CNG']
print(f"Does the database include any CNG vehicle? {'Yes' if not cng_vehicles.empty else 'No'}. If yes, there are {len(cng_vehicles)} CNG vehicles.")
```

```python
# Determine the number of vehicles for sale from individuals
individual_sales = data[data['Seller_Type'] == 'Individual']
print(f"There are {len(individual_sales)} vehicles for sale from individuals directly.")
```

```python
# Check for automatic transmission vehicles
auto_transmission_vehicles = data[data['Transmission'] == 'Automatic']
print(f"Does this database contain auto transmission vehicles? {'Yes' if not auto_transmission_vehicles.empty else 'No'}. If yes, there are {len(auto_transmission_vehicles)} auto transmission vehicles.")
```

```python
# Determine the number of single person owned vehicles
single_person_owned = data[data['Owner'] == 1]
print(f"There are {len(single_person_owned)} single person owned vehicles in this database.")
```

```python
# Determine the most and least cost depreciated vehicle
data['Depreciation'] = data['Present_Price'] - data['Selling_Price']
most_depreciated_vehicle = data.loc[data['Depreciation'].idxmax()]
least_depreciated_vehicle = data.loc[data['Depreciation'].idxmin()]
print(f"The most cost depreciated vehicle is {most_depreciated_vehicle['Car_Name']} with a depreciation of {most_depreciated_vehicle['Depreciation']}.")
print(f"The least cost depreciated vehicle is {least_depreciated_vehicle['Car_Name']} with a depreciation of {least_depreciated_vehicle['Depreciation']}.")
```

```python
# Determine the brands of vehicles less affected by cost depreciation
brands_depreciation = data.groupby('Car_Name')['Depreciation'].mean().sort_values()
print(f"Brands of vehicles less affected by cost depreciation: {brands_depreciation.head()}.")
```

```python
# Factors affecting cost depreciation
numeric_data = data.select_dtypes(include=[np.number])
correlation = numeric_data.corr()
print(correlation['Depreciation'])
```

```python
# Check if selling price is affected by age of vehicle and distance driven
numeric_data = data.select_dtypes(include=[np.number])
correlation = numeric_data.corr()
print(correlation['Selling_Price'])
```

```python
newest_vehicles = data[data['Year'] > 2014]
print(newest_vehicles)
```

```python
# Filter two-wheelers based on 'Car_Name' containing specific names
two_wheelers = data[data['Car_Name'].str.contains('Activa|Jupyter|Access|Wego', case=False, na=False)]
print(two_wheelers)
```

```python
# Determine the oldest bike sold
oldest_bike = two_wheelers.loc[two_wheelers['Year'].idxmin()]
print(f"The oldest bike sold is {oldest_bike['Car_Name']} from the year {oldest_bike['Year']}.")
```

```python
# Determine the newest bike sold
newest_bike = two_wheelers.loc[two_wheelers['Year'].idxmax()]
print(f"The newest bike sold is {newest_bike['Car_Name']} from the year {newest_bike['Year']}.")
```

```python
# Determine the most sold bike
most_sold_bike = two_wheelers['Car_Name'].value_counts().idxmax()
print(f"The most sold bike in this data is {most_sold_bike}.")
```
